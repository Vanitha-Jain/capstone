```python

data['bioactivity_class'].replace('active', 'P', inplace=True)
data['bioactivity_class'].replace('inactive', 'N', inplace=True)
data.rename(columns = {"class": "class"}, inplace = True)
data
```

```python
data['bioactivity_class'].replace('P', '1', inplace=True)
data['bioactivity_class'].replace('N', '0', inplace=True)
data.rename(columns = {"class": "class"}, inplace = True)
data
```

```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
import lazypredict
from lazypredict.Supervised import LazyRegressor
```

```python
!pip install dask[dataframe]
```

```python
X = data.drop('bioactivity_class', axis=1)
Y = data.bioactivity_class
```

```python
X.shape
```

```python

from sklearn.feature_selection import VarianceThreshold
selection = VarianceThreshold(threshold=(.8 * (1 - .8)))
X = selection.fit_transform(X)
X.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
clf = LazyRegressor(verbose=0,ignore_warnings=True, custom_metric=None)
models_train,predictions_train = clf.fit(X_train, X_train, Y_train, Y_train)
models_test,predictions_test = clf.fit(X_train, X_test, Y_train, Y_test)
```
