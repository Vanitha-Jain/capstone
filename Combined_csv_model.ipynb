```python
import pandas as pd

# Load the uploaded CSV files
file_1_path = '/content/Train_Combined.csv'
file_2_path = '/content/Test_Combined.csv.csv'

# Read the files into dataframes
df1 = pd.read_csv(file_1_path)
df2 = pd.read_csv(file_2_path)

# Reorder columns of df2 to match df1
df2 = df2[df1.columns]

# Combine the dataframes
combined_df = pd.concat([df1, df2], ignore_index=True)

# Save the combined dataframe
output_path = '/content/combined_data_model.csv'
combined_df.to_csv(output_path, index=False)

output_path
```
